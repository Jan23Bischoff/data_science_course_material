```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats

diamonds = sns.load_dataset("diamonds")
diamonds_sample = diamonds.sample(1000, random_state=42)
```

# Solution Exercise 1

```python
sns.set_theme(style="whitegrid") # Setting the theme
sns.histplot(data=diamonds, x="price")
plt.show()
```

# Solution Exercise 2

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=diamonds_sample, x="carat", y="price")
sns.despine() # 1. Remove spines
plt.title("Diamond Price vs Weight") # 2. Clear Title
plt.xlabel("Carat (Weight)") # 3. Units
plt.ylabel("Price (USD)")    # 3. Units
plt.show()
```

# Solution Exercise 4

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=diamonds_sample,
    x="carat", y="price",
    hue="cut", 
    size="depth",
    alpha=0.6,      # Transparency
    palette="deep"
)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2) # Move legend outside
plt.show()
```

# Solution Exercise 5

```python

sns.relplot(
    data=diamonds_sample,
    x="carat", y="price",
    col="cut",      # Faceting
    kind="scatter"
)
plt.show()
```

# Solution Exercise 6

```python
sns.relplot(
    data=diamonds_sample,
    x="carat", y="price",
    col="cut",
    col_wrap=3,          # Wrap columns
    hue="cut",           # Optional: color by cut too for clarity
    palette="viridis",   # Colorblind friendly palette
    kind="scatter"
)
plt.show()
```

# Solution Exercise 7

```python
plt.figure(figsize=(10, 6))
# Boxen plot handles large datasets/heavy tails better than boxplot
sns.boxenplot(data=diamonds, x="color", y="price", palette="Set2")
plt.title("Price Distribution by Diamond Color (Boxen Plot)")
plt.show()
```

# Solution Exercise 8 (Raincloud)

```python
f, ax = plt.subplots(figsize=(12, 6))

# 1. Violin
sns.violinplot(data=diamonds_sample, x="cut", y="price", 
               color=".9", inner=None, ax=ax)

# 2. Strip
sns.stripplot(data=diamonds_sample, x="cut", y="price", 
              color="blue", alpha=0.3, jitter=True, ax=ax)

# 3. Box (Optional overlay)
sns.boxplot(data=diamonds_sample, x="cut", y="price", 
            width=0.2, boxprops={'zorder': 2, 'facecolor':'none'}, ax=ax)

ax.set_title("Raincloud Plot: Price by Cut")
plt.show()
```

# Solution Exercise 9 (Normality Check)

```python
fig, ax = plt.subplots(figsize=(6, 6))
# Q-Q plot compares data quantiles to theoretical normal quantiles
stats.probplot(diamonds['carat'], plot=ax)
ax.set_title("Q-Q Plot for Carat")
plt.show()
# Interpretation: It deviates significantly from the red line, 
# so 'carat' is NOT normally distributed.
```

# Solution Exercise 10 (Regression)

```python
plt.figure(figsize=(8, 6))
sns.regplot(
    data=diamonds_sample, 
    x="carat", 
    y="price",
    scatter_kws={'alpha': 0.2, 'color': 'blue'}, # Transparent points
    line_kws={'color': 'red'}                    # Contrasting line
)
plt.title("Linear Regression: Price vs Carat")
plt.show()
```
